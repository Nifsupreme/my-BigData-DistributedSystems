## Create dataframes
b = common friend of (a,c)
ab = bc = 2-way edgelist

```python
ab = {
    'A': [1,2,1,3,1,4,2,3,2,5,3,4,3,5,3,6,3,7],
    'B': [2,1,3,1,4,1,3,2,5,2,4,3,5,3,6,3,7,3]
}

ab_df = pd.DataFrame(ab)
ab_df.head()
```

```python
bc = {
    'B': [1,2,1,3,1,4,2,3,2,5,3,4,3,5,3,6,3,7],
    'C': [2,1,3,1,4,1,3,2,5,2,4,3,5,3,6,3,7,3]
}

bc_df = pd.DataFrame(bc)
bc_df.head()
```

## Join

```python
abc_df = pd.merge(ab_df, bc_df, left_on='B', right_on='B')
abc_df.head()
```

## Count the number of common friends

```python
df = abc_df.groupby(['A', 'C']).count()
df
```
